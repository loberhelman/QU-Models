```python
# if running this code in a Jupter notbeook or JupyterLab
%matplotlib ipympl
import mpl_interactions.ipyplot as iplt
import matplotlib.pyplot as plt
from ipywidgets import IntSlider, Label

import numpy as np
c = 299792458.

psi_0 = 95
frac_pol = np.linspace(0.01, 0.5, 100)
RM_src = np.linspace (1, 101, 100)
RM_screen = np.linspace (-1100, 1100, 100)

#x axis
lower_freq_lim_Hz = 800.e6
upper_freq_lim_Hz = 1089.e6
channel_width_Hz = 1.e6
freq_array = np.arange(lower_freq_lim_Hz, upper_freq_lim_Hz, channel_width_Hz)
lambda_sqr_array = np.power(c/freq_array, 2.0)

#def add_noise_SNR 

def real_slab(lambda_sqr_array, frac_pol, RM_screen, RM_src):
    return (frac_pol * np.exp(2j * (np.radians(psi_0) + (0.5 * RM_src + RM_screen) * lambda_sqr_array)) * ((np.sin(RM_src * lambda_sqr_array)) / (RM_src * lambda_sqr_array))).real 

def imag_slab(lambda_sqr_array, frac_pol, RM_screen, RM_src):
    return (frac_pol * np.exp(2j * (np.radians(psi_0) + (0.5 * RM_src + RM_screen) * lambda_sqr_array)) * ((np.sin(RM_src * lambda_sqr_array)) / (RM_src * lambda_sqr_array))).imag 

def slab(lambda_sqr_array, frac_pol, RM_screen,RM_src):
    imaginary =  imag_slab(lambda_sqr_array, frac_pol, RM_screen,RM_src)
    real = real_slab(lambda_sqr_array, frac_pol, RM_screen,RM_src)
    return np.sqrt(real**2 + imaginary**2)

fig, ax = plt.subplots(figsize=[4,4])
fig.canvas.header_visible = False

controls = iplt.plot(lambda_sqr_array, real_slab, frac_pol=frac_pol, RM_screen=RM_screen, RM_src=RM_src, color = 'blue',label="Q", linewidth=3)
iplt.plot(lambda_sqr_array, imag_slab, controls=controls,color='red', label="U", linewidth=3)
iplt.plot(lambda_sqr_array, slab , controls=controls,color='black', label="P",linewidth=3)
_ = plt.legend(loc='upper right')

iplt.xlabel('$\lambda^2$', fontsize=18)
iplt.ylabel('P($\lambda^2$)', fontsize=18)
iplt.title('Uniform Slab')
plt.tight_layout()
plt.show()
```
