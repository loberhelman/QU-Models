## This notebook is for exploring the SNR of the QU simulations 

# I only ran a simple 10 sources with the same parameters but different SNR

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


#Load condensed json file
slab_10_QUfit_df = pd.read_csv('/Users/loberhel/Desktop/Simulated_sources_code/Sim_tests/SNR_testing/condensed_sim_burn_slab_model_num_5_sims.csv')
slab_10_true_df = pd.read_csv('/Users/loberhel/Desktop/Simulated_sources_code/Sim_tests/SNR_testing/burnslab_number_10_freq_800_1089_ran_SNR.csv')

FS_10_QUfit_df = pd.read_csv('/Users/loberhel/Desktop/Simulated_sources_code/Sim_tests/SNR_testing/simfaraday_simple_model_num_1_sims.csv')
FS_10_true_df = pd.read_csv('/Users/loberhel/Desktop/Simulated_sources_code/Sim_tests/SNR_testing/faraday_simple_number_10_freq_800_1089_ran_SNR.csv')
print(FS_10_true_df.SNR)



FS_SNR = FS_10_QUfit_df.fracPol / (FS_10_QUfit_df.fracPol_plus_err)

SNR_estimate = slab_10_QUfit_df.fracPol / (slab_10_QUfit_df.fracPol_minus_err)
#print(slab_10_true_df['RM_src_radm2'])
```

```python
plt.plot(slab_10_true_df.SNR, SNR_estimate, 'o', label='slab')
plt.plot(FS_10_true_df.SNR, FS_SNR, 'o', label='FS')
plt.legend()
plt.xlabel('True SNR')
plt.ylabel('SNR QU fit estimate')

```

```python
residual_RM = slab_10_QUfit_df['deltaRM_radm2']  - slab_10_true_df['RM_src_radm2']
slab_10_true_df.keys()
plt.errorbar(slab_10_true_df['SNR'], residual_RM, np.array([slab_10_QUfit_df.deltaRM_radm2_plus_err,  slab_10_QUfit_df.deltaRM_radm2_minus_err]), fmt='r^', label='$\Delta$ RM')
plt.ylabel('$residual')
plt.xlabel('SNR')
plt.title('10 Unif Slab')



residual_fracpol = slab_10_QUfit_df['fracPol']  - slab_10_true_df['frac_pol']
slab_10_true_df.keys()
plt.errorbar(slab_10_true_df['SNR'], residual_fracpol, np.array([slab_10_QUfit_df.fracPol_plus_err,  slab_10_QUfit_df.fracPol_minus_err]), fmt='b^', label='frac_pol')
plt.ylabel('residual')
plt.xlabel('SNR')
plt.title('10 Unif Slab')


residual_polangle = slab_10_QUfit_df['psi0_deg']  - slab_10_true_df['psi0_deg']
slab_10_true_df.keys()
plt.errorbar(slab_10_true_df['SNR'], residual_polangle, np.array([slab_10_QUfit_df.psi0_deg_plus_err,  slab_10_QUfit_df.psi0_deg_minus_err]), fmt='g^', label='$\psi_0$')
plt.ylabel('residual')
plt.xlabel('SNR')
plt.title('10 Unif Slab')

#plt.yscale('log')


plt.legend()
```
